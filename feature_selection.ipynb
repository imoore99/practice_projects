```python
#https://www.datacamp.com/tutorial/feature-selection-python

import pandas as pd
import numpy as np

# Import the necessary libraries first
from sklearn.feature_selection import SelectKBest, chi2, RFE
from sklearn.linear_model import LogisticRegression, Ridge
```

```python
# load data
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
df = pd.read_csv(url, names=names)
```

```python
df.describe()
```

```python
array = df.values
X = array[:,0:8]
Y = array[:,8]
```

Chi-Squared Test

```python
# Feature extraction
test = SelectKBest(score_func=chi2, k=4)
fit = test.fit(X, Y)

# Summarize scores
np.set_printoptions(precision=3)
print(fit.scores_)

features = fit.transform(X)
# Summarize selected features
print(features[0:5,:])

cols_idxs = fit.get_support(indices=True)
features_df_new = df.iloc[:, cols_idxs]
features_df_new.columns #Columns with highest chi-squared score

```

Recursive Feature Elimination

```python
# Feature extraction
model = LogisticRegression()
rfe = RFE(model, n_features_to_select=3, step = 1)
fit = rfe.fit(X, Y)
print("Num Features: %s" % (fit.n_features_))
print("Selected Features: %s" % (fit.support_))
print("Feature Ranking: %s" % (fit.ranking_))

idxs = fit.get_support(indices=True)
features_df_new = df.iloc[:, idxs]
features_df_new.columns
```

Ridge Regression

```python
ridge = Ridge(alpha=1.0)
ridge.fit(X,Y)

Ridge(alpha=1.0, copy_X=True, fit_intercept=True, max_iter=None,
   random_state=None, solver='auto', tol=0.001)
```

```python
# A helper method for pretty-printing the coefficients
def pretty_print_coefs(coefs, names = None, sort = False):
    if names == None:
        names = ["X%s" % x for x in range(len(coefs))]
    lst = zip(coefs, names)
    if sort:
        lst = sorted(lst,  key = lambda x:-np.abs(x[0]))
    return " + ".join("%s * %s" % (round(coef, 3), name)
                                   for coef, name in lst)
```

```python
print ("Ridge model:", pretty_print_coefs(ridge.coef_))
```
